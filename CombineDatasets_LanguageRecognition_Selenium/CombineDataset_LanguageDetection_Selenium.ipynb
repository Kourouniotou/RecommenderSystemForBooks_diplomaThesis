### Selenium on Goodreads site

```python
import csv
from getpass import getpass
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver import Chrome
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException   
import numpy as np
from selenium.webdriver.common.by import By


driver = Chrome("C:\\Users\\antonia\\Downloads\\chromedriver.exe")
driver.get("https://www.goodreads.com/user/sign_in")
driver.find_element(By.LINK_TEXT, "Sign in with email").click()
username = driver.find_element(By.XPATH,"//input[@name='email']")
username.send_keys("email")


my_password = "password"
password = driver.find_element(By.XPATH, "//input[@name='password']")
password.send_keys(my_password)

password.send_keys(Keys.RETURN)

def check_exists_by_xpath(xpath):
    try:
        driver.find_element(By.XPATH,xpath)
    except NoSuchElementException:
        return False
    return True

def getDescriptionSelenium(isbn, driver):
    check = 1
    
    search_input = driver.find_element(By.XPATH, '//input[@class = "searchBox__input searchBox__input--navbar" ]')
        
    search_input.send_keys(isbn)

    search_input.send_keys(Keys.RETURN)
         
    try:
        link = driver.find_element(By.LINK_TEXT, '...more')
    except NoSuchElementException:
        check = 0
    if(check == 1):
        ActionChains(driver).move_to_element(link).perform()
        link.click()
        if(check_exists_by_xpath("//div[@id='description']")):
            description = link.find_elements(By.XPATH, "//div[@id='description']")[0].text
            description = description.replace(" (less)", "")
        else:
            description = np.nan
    elif(check == 0):
        if(check_exists_by_xpath("//div[@id='description']")):
            description = driver.find_elements(By.XPATH, "//div[@id='description']")[0].text
        else:
            description = np.nan
    return(description)
```
